```python
new_data =  data.sort_values(by=['rentrée'])
#new_data["rentrée"].max()
new_data.columns.tolist()
new_data_sum = pd.concat((new_data['rentrée'],new_data['Sexe de l’étudiant'],new_data['Nombre total d’étudiants inscrits']),
                         axis=1,keys= ['rentrée','Sexe de l’étudiant','Nombre total d’étudiants inscrits'])
new_data_sum
```

```python

#enlever les emptys rows dans la column Nombre Total Etudiants
new_data_sum=new_data_sum.dropna()
column_name_nbrTotal=new_data_sum.columns[new_data_sum.columns.str.startswith('Nombre total')][0]
#new_data_sum.iloc[:,2:3].sum()
column_name_nbrTotal


#data[column_name].sum()
```

```python
column_name_sexe=new_data_sum.columns.tolist()[1]
column_name_annee=new_data_sum.columns.tolist()[0]
nbr_total_feminin=new_data_sum[new_data_sum[column_name_sexe].isin(['Feminin'])][column_name_nbrTotal].sum()
print("nbr total feminin est  : ",nbr_total_feminin)
print("\nnbr total feminin selon l'année : ")
dataframe_global=new_data_sum[new_data_sum[column_name_sexe].isin(['Feminin'])].groupby(new_data_sum[column_name_annee],axis=0)[column_name_nbrTotal].sum().reset_index(name ='Nombre Etudiant Femme')
#new_data_sum.groupby(data["rentrée"],axis=0)
dataframe_global
```

```python
nbr_total_masculin=new_data_sum[new_data_sum[column_name_sexe].isin(['Masculin'])][column_name_nbrTotal].sum()
print("nbr total masculin est  : ",nbr_total_masculin)
print("\nnbr total masculin selon l'année : ")
dataframe_male=new_data_sum[new_data_sum[column_name_sexe].isin(['Masculin'])].groupby(new_data_sum[column_name_annee],axis=0)[column_name_nbrTotal].sum().reset_index(name ='Nombre Etudiant Male')

dataframe_global["Nombre Etudiant Male"]=dataframe_male["Nombre Etudiant Male"]

dataframe_global
```
