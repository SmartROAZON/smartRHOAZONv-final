```python
#imports
import pandas as pd
import numpy as np
from sklearn import cluster
```

```python
data = pd.read_csv("data/lotissements.csv",sep=';') 

data.info()
```

```python
data.head()
```

```python
new_data =  data.sort_values(by=['objectid'])
#new_data["rentrée"].max()
new_data.columns.tolist()
new_data_sum = pd.concat((new_data['objectid'],new_data['vocation'],new_data['etape'],new_data['datesig'],new_data['nomlot']),
                         axis=1)
```

```python
new_data_sum.count()
```

```python
new_data_sum
```

```python
#new_data_sum['datesig'] = pd.to_datetime(new_data_sum['datesig'], format='%d-%m-%Y')
```

```python
new_data_sum['datesig']=new_data_sum['datesig'].str.slice(start=0, stop=8)
new_data_sum["year"]=new_data_sum['datesig'].str.slice(start=0, stop=4)
new_data_sum["month"]=new_data_sum['datesig'].str.slice(start=4, stop=6)
new_data_sum["day"]=new_data_sum['datesig'].str.slice(start=6, stop=8)
new_data_sum
groupbyyear=new_data_sum['year'].groupby(new_data_sum['year'],axis=0).count().reset_index(name="NumberConstructionPeerYear")
groupbyyear
```
