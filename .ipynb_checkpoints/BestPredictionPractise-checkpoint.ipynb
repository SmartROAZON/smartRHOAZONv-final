```python
# Afin de faire les prédictions dans le future avec la régression linéaire on a besoin d'abord de savoir nos "features"
# ( variable qui aident à la prédiction) c'est à dire :
# pour une équation de regression lineaire :
# y = a * x1 + b * x2 + c 
# on a besoin de x1 et x2 afin de pouvoir prédire notre y
# Donc la notion de temps est inhexistante dans cette equation afin de palier à ce problème, on peut imaginer ceci :
# On prédit d'abord nos variables qui nous aideront à la prédiction 
# En d'autre termes, prédire x1 et x2 dans le future, et utiliser ces valeurs pour prédire notre y
# Du coup on aurait une prédiction dans le futur de y
# Pour la prédiction de y on utiliserait la régression linéaire
# Pour x1 et x2 on doit utiliser la régression mais uniquement en se basant sur le temps (times series)
# c'est à dire pour prédire x1, on se basera sur le temps et donc l'équation de prédiction de x1 et x2 seraient en gros :
#       y = a * (temps) + b
#   y est notre x1 ou x2
#   temps serait le "feature" et b "l'intercept"
```

```python
def highlight_greaterthan(s,threshold,column):
    is_max = pd.Series(data=False, index=s.index)
    is_max[column] = s.loc[column] >= threshold
    #print(type(is_max))
    return ['background-color: yellow' if is_max.any() else '' for v in is_max]


#data.style.apply(highlight_max)
data.style.apply(highlight_greaterthan,threshold=2016,column=['rentrée'], axis=1)
d = pd.concat([data_etudiant,data], ignore_index=True)
d.style.apply(highlight_greaterthan,threshold=2016,column=['rentrée'], axis=1)
```

# Arima model

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
#p,d,q  p = periods taken for autoregressive model
#d -> Integrated order, difference
# q periods in moving average model
model_arima = ARIMA(train,order=(0, 1, 1))
model_arima_fit = model_arima.fit()
print(model_arima_fit.aic)
```

```python
pred= model_arima_fit.forecast(steps=5)[0]
pred
```

```python
plt.plot(test)
plt.plot(pred,color='red')
```

```python
mean_squared_error(test,pred)
```

```python
import itertools
p=d=q=range(0,10)
pdq = list(itertools.product(p,d,q))
pdq
```

```python
import warnings
warnings.filterwarnings('ignore')
min_aic = 9999
for param in pdq:
    try:
        model_arima = ARIMA(train,order=param)
        model_arima_fit = model_arima.fit()
        #print(param,model_arima_fit.aic)
        if model_arima_fit.aic < min_aic:
            min_aic=model_arima_fit.aic
            comb=param
        
    except:
        continue
        
print(min_aic,comb)
```
